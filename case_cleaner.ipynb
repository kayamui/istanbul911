```python
#This function is for to extract external stations from the total stations to find internal(governmental) stations
def clearData():
  import pandas as pd

  # Import all notices
  df_big = pd.read_excel('/content/İhbar işlemleri (75).xls')
  # Import external notices
  df_small = pd.read_excel('/content/İhbar işlemleri (76).xls')

  df_big[~df_big['Ekip Kodu'].isin(df_small['Ekip Kodu'])].sort_values(by = 'KKM Protokol').to_excel('CleanCases.xlsx')
clearData()
```
