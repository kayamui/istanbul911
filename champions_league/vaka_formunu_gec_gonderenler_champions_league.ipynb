```python
import pandas as pd
import datetime
```

```python
df1 = pd.read_excel('/content/Personel-Nöbet-Listesi (1).xls')
df2 = pd.read_excel('/content/Personel-Nöbet-Listesi (2).xls')
df3 = pd.read_excel('/content/Personel-Nöbet-Listesi (3).xls')
df4 = pd.read_excel('/content/Personel-Nöbet-Listesi (4).xls')
```

```python
df_merged = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df_merged.dropna(axis=1,how= 'all', inplace = True)
```

```python
df_merged.fillna('False', inplace = True)
```

```python
df_merged[(df_merged['Unnamed: 1'] != 'False') & (df_merged['Unnamed: 2'] == 'False')]
```

```python
df_merged = df_merged[(df_merged['Unnamed: 1'] != 'False') | (df_merged['Unnamed: 2'] != 'False')]
```

```python
team_codes = []
team_code = ''
for i in range(len(df_merged)):
  if ((df_merged.iloc[i]['Unnamed: 1'] != 'False') & (df_merged.iloc[i]['Unnamed: 2'] == 'False')):
    team_code = df_merged.iloc[i]['Unnamed: 1']
  team_codes.append(team_code)
```

```python
len(team_codes)
```

```python
len(df_merged)
```

```python
df_merged['Ekip No'] = team_codes
```

```python
df_merged.reset_index(drop= True, inplace= True)
```

```python
df_merged.columns = df_merged.iloc[1]
```

```python
df_merged = df_merged[df_merged['Soyisim'] != 'False']
```

```python
df_merged = df_merged[df_merged['İsim'] != 'İsim']
```

```python
df_merged.rename(columns = {'.ASKOM ARNAVUTKÖY':'Ekip No'}, inplace = True)
```

```python
df_merged
```

```python
staff_list = df_merged[['İsim', 'Soyisim', 'Kimlik No', 'Kadro', 'Telefon', 'Başlangıç Tarihi', 'Bitiş Tarihi', 'Ekip No']]
```
